```python
## Splitting document into smaller chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_documents(documents, chunk_size=1000, chunk_overlap=200):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators= ["\n\n", "\n", " ", ""]
    )

    split_docs = text_splitter.split_documents(documents)
    print(f"Split {len(documents)} documents into {len(split_docs)} chunks")

    # Show example chunk
    if split_docs:
        print("\nExample chunk:")
        print("Content: ", split_docs[0].page_content[:500])  # Print first 500 characters of the first chunk
        print("Metadata: ", split_docs[0].metadata)

    return split_docs
```

## Embedding and VectorStore DB


```python
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import uuid
from typing import List, Dict, Any
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class EmbeddingManager:
    """
    Handles embedding generation using SentenceTransformer.
    """

    def __init__(self, model_name: str = "all-MiniLM-L6-v2"):
        """Initializes the EmbeddingManager with a specific model.

        Args:
            model_name (str, optional): The name of the SentenceTransformer model to use. Defaults to "all-MiniLM-L6-v2".
        """
        self.model_name = model_name
        self.model = None
        self.load_model()

    def load_model(self):
        """Loads the SentenceTransformer model."""
        try:
            print(f"Loading Embedding Model: {self.model_name}")
            self.model = SentenceTransformer(self.model_name)
            print("Model loaded successfully. Embedding dimension is", self.model.get_sentence_embedding_dimension())
        except Exception as e:
            print(f"Error loading model {self.model_name}: {e}")
            raise

    def generate_embeddings(self, texts: List[str]) -> np.ndarray:
        """Generates an embedding for a list of texts.

        Args:
            texts (List[str]): The input texts to embed.

        Returns:
            np.ndarray: Numpy array of embeddings with shape (len(texts), embedding_dimension).    
        """
        try:
            if not self.model:
                raise ValueError("Model is not loaded.")
            
            print(f"Generating embedding for {len(texts)} texts")
            embeddings = self.model.encode(texts, show_progress_bar=True)
            print("Embedding generated with shape:", embeddings.shape)
            return embeddings
        except Exception as e:
            print(f"Error generating embedding for texts: {e}")
            return None
        

## Initialize the embedding manager
embedding_manager = EmbeddingManager(model_name="all-MiniLM-L6-v2")
embedding_manager
```

## VectorStore

```python
class VectorStore:
    """
    Manages a vector store for embeddings using ChromaDB.
    """
    def __init__(self, collection_name: str = "pdf_documents", persist_directory: str = "../data/VectorStore/DB"):
        """Initializes the VectorStore with a ChromaDB client and collection.

        Args:
            collection_name (str, optional): The name of the ChromaDB collection. Defaults to "pdf_documents".
            persist_directory (str, optional): The directory to persist the vector store. Defaults to "../data/VectorStore/DB".
        """
        self.collection_name = collection_name
        self.persist_directory = persist_directory
        self.client = None
        self.collection = None
        self.__initialize_store()

    def __initialize_store(self):
        """Initializes the ChromaDB client and collection."""
        try:
            print("Initializing ChromaDB client and collection...")

            # create a persistent ChromaDB client
            os.makedirs(self.persist_directory, exist_ok=True)
            self.client = chromadb.PersistentClient(path=self.persist_directory)

            # get or create the collection
            self.collection = self.client.get_or_create_collection(
                name=self.collection_name,
                metadata={"description": "Collection for storing pdf document embeddings"}
            )
            print(f"ChromaDB client and collection '{self.collection_name}' initialized.")
            print(f"Existing collections: {self.collection.count()}")
        except Exception as e:
            print(f"Error initializing ChromaDB client or collection: {e}")
            raise
        

    def initialize_client(self):
        """Initializes the ChromaDB client."""
        try:
            print("Initializing ChromaDB client...")
            self.client = chromadb.Client(Settings(
                chroma_db_impl="duckdb+parquet",
                persist_directory=".chromadb"
            ))
            print("ChromaDB client initialized.")
        except Exception as e:
            print(f"Error initializing ChromaDB client: {e}")
            raise

    def add_documents(self, documents: List[Any], embeddings: np.ndarray):
        """Adds documents and their embeddings to the vector store.

        Args:
            documents (List[Any]): The input documents.
            embeddings (np.ndarray): The corresponding embeddings.
        """
        if len(documents) != len(embeddings):
            raise ValueError("Number of documents must match number of embeddings.")
        
        print(f"Adding {len(documents)} documents to the vector store...")

        # Prepare data for vector store
        ids = []
        metadatas = []
        documents_text = []
        embeddings_list = []

        for i, (doc, emb) in enumerate(zip(documents, embeddings)):
            # generate a unique ID for each document
            doc_id = str(uuid.uuid4())
            ids.append(doc_id)

            # prepare metadata
            metadata = dict(doc.metadata)
            metadata['doc_index'] = i
            metadata['context_length'] = len(doc.page_content) 
            metadatas.append(metadata)

            # Document content
            documents_text.append(doc.page_content if hasattr(doc, 'page_content') else str(doc))

            # Embedding as list
            embeddings_list.append(emb.tolist())
        
        # Add to collection
        try:
            self.collection.add(
                ids=ids,
                documents=documents_text,
                embeddings=embeddings_list,
                metadatas=metadatas
            )
            print(f"Successfully added {len(documents)} documents to the collection.")
            print(f"Total documents in collection: {self.collection.count()}")
        
        except Exception as e:
            print(f"Error adding documents to the collection: {e}")
            raise

    def add_documents_batch(self, documents, batch_size=100):
        for i in range(0, len(documents), batch_size):
            batch = documents[i:i+batch_size]
            batch_embeddings = self.embedding_manager.generate_embeddings(
                [doc.page_content for doc in batch]
            )
            self.vector_store.add_documents(batch, batch_embeddings)

    def add_new_documents_safely(self, new_documents):
        # Check for duplicates (optional)
        existing_count = self.vector_store.collection.count()
        print(f"Existing documents: {existing_count}")
        
        # Process in batches
        self.add_documents_batch(new_documents, batch_size=100)
        
        # Verify addition
        new_count = self.vector_store.collection.count()
        print(f"Added {new_count - existing_count} new documents")
        print(f"Total documents: {new_count}")

# Initialize the vector store
vector_store = VectorStore(collection_name="pdf_documents", persist_directory="../data/VectorStore/DB")
vector_store
```

## Retriver Pipeline From Vectorstore


```python
class RAGRetriever:
    """Retriver Pipeline From Vectorstore"""
    def __init__(self, vector_store: VectorStore, embedding_manager: EmbeddingManager):
        """Initializes the RAGRetriever with a vector store and embedding manager.

        Args:
            vector_store (VectorStore): The vector store instance.
            embedding_manager (EmbeddingManager): The embedding manager instance.
        """
        self.vector_store = vector_store
        self.embedding_manager = embedding_manager

    def retrieve(self, query: str, top_k: int = 5, score_threshold: float = 0.0) -> List[Dict[str, Any]]:
        """Retrieves the most relevant documents for a given query.

        Args:
            query (str): The input query string.
            top_k (int, optional): The number of top documents to retrieve. Defaults to 5.
            score_threshold (float, optional): The minimum similarity score to consider a document relevant. Defaults
                to 0.5.
        Returns:
            List[Dict[str, Any]]: A list of retrieved documents with their metadata and similarity scores.
        """
        print(f"Retrieving documents for query: '{query}'")
        print(f"Top_k: {top_k}, Score Threshold: {score_threshold}")

        # Generate embedding for the query
        query_embedding = self.embedding_manager.generate_embeddings([query])[0]
        if query_embedding is None:
            print("Error generating query embedding.")
            return []

        try:
            # Retrieve similar documents from the vector store
            results = self.vector_store.collection.query(
                query_embeddings=[query_embedding.tolist()],
                n_results=top_k
            )

            # Process results
            retrieved_docs = []

            if results['documents'] and results['documents'][0]:
                documents = results['documents'][0]
                metadatas = results['metadatas'][0]
                distances = results['distances'][0]
                ids = results['ids'][0]
                
                for i, (doc_id, doc, metadata, distance) in enumerate(zip(ids, documents, metadatas, distances)):
                    # Covert distance to similarity score
                    similarity_score = 1 - distance  # Assuming distance is in [0, 1]

                    if similarity_score >= score_threshold:
                        retrieved_docs.append({
                            "id:": doc_id,
                            "document": doc,
                            "metadata": metadata,
                            "similarity_score": similarity_score,
                            "distance": distance,
                            "rank": i + 1
                        })

                print(f"Retrieved {len(retrieved_docs)} documents above the score threshold {score_threshold}.")
            else:
                print("No documents found in the vector store.")

            return retrieved_docs
        except Exception as e:
            print(f"Error retrieving documents: {e}")
            return []

rag_retriever = RAGRetriever(vector_store=vector_store, embedding_manager=embedding_manager)
```

```python
# Enhanced query processing function
async def enhanced_rag_comparison(query, use_both=True):
    """
    Compare traditional RAG vs Agentic RAG responses
    """
    results = {
        'query': query,
        'traditional': None,
        'agentic': None,
        'comparison': {}
    }
    
    print(f"🔍 Processing Query: {query}")
    print("=" * 80)
    
    # Traditional RAG
    if use_both:
        print("🔄 Running Traditional RAG...")
        traditional_start = time.time()
        try:
            traditional_response = simple_rag(
                query=query, 
                retriever=rag_retriever, 
                llm=llm, 
                top_k=3
            )
            traditional_time = time.time() - traditional_start
            
            results['traditional'] = {
                'answer': traditional_response,
                'processing_time': traditional_time,
                'method': 'Simple retrieval + LLM generation'
            }
            print(f"✅ Traditional RAG completed in {traditional_time:.2f}s")
        except Exception as e:
            print(f"❌ Traditional RAG failed: {e}")
            results['traditional'] = {'error': str(e)}
    
    # Agentic RAG
    if agentic_rag:
        print("🔄 Running Agentic RAG...")
        agentic_start = time.time()
        try:
            agentic_response = await agentic_rag.process_query(query)
            agentic_time = time.time() - agentic_start
            
            results['agentic'] = {
                'answer': agentic_response['answer'],
                'confidence': agentic_response['confidence'],
                'insights': agentic_response['insights'],
                'quality_score': agentic_response['quality_assessment']['score'],
                'processing_time': agentic_time,
                'agents_used': agentic_response['processing_metadata']['agents_used'],
                'validation': agentic_response['processing_metadata'].get('validation', {}),
                'sources_count': len(agentic_response['sources'])
            }
            print(f"✅ Agentic RAG completed in {agentic_time:.2f}s")
        except Exception as e:
            print(f"❌ Agentic RAG failed: {e}")
            results['agentic'] = {'error': str(e)}
    
    return results

# Test the enhanced system
test_query = "Tell me about Virtual Drawing using Machine Learning"
```

## Performance Analysis and System Monitoring

Monitor the performance and behavior of both traditional and agentic RAG systems.

```python
# Performance analysis function
def analyze_rag_performance(results_list):
    """
    Analyze performance metrics across multiple queries
    """
    if not results_list:
        return "No results to analyze"
    
    # Extract metrics
    traditional_times = []
    agentic_times = []
    quality_scores = []
    confidence_levels = []
    
    for result in results_list:
        if result.get('traditional') and 'processing_time' in result['traditional']:
            traditional_times.append(result['traditional']['processing_time'])
        
        if result.get('agentic') and 'processing_time' in result['agentic']:
            agentic_times.append(result['agentic']['processing_time'])
            
        if result.get('agentic') and 'quality_score' in result['agentic']:
            quality_scores.append(result['agentic']['quality_score'])
            
        if result.get('agentic') and 'confidence' in result['agentic']:
            confidence_levels.append(result['agentic']['confidence'])
    
    # Calculate statistics
    analysis = {
        'traditional_avg_time': sum(traditional_times) / len(traditional_times) if traditional_times else 0,
        'agentic_avg_time': sum(agentic_times) / len(agentic_times) if agentic_times else 0,
        'avg_quality_score': sum(quality_scores) / len(quality_scores) if quality_scores else 0,
        'confidence_distribution': {
            'high': confidence_levels.count('high'),
            'medium': confidence_levels.count('medium'),
            'low': confidence_levels.count('low')
        }
    }
    
    return analysis

# System health check
def check_system_health():
    """Check the health of both RAG systems"""
    health_report = {
        'traditional_rag': {
            'vector_store': 'unknown',
            'embedding_manager': 'unknown',
            'llm': 'unknown'
        },
        'agentic_rag': {
            'database': 'unknown',
            'agents': 'unknown',
            'integration': 'unknown'
        }
    }
    
    # Check traditional RAG components
    try:
        doc_count = vector_store.collection.count()
        health_report['traditional_rag']['vector_store'] = f'✅ Healthy ({doc_count} documents)'
    except Exception as e:
        health_report['traditional_rag']['vector_store'] = f'❌ Error: {str(e)[:50]}'
    
    try:
        test_embedding = embedding_manager.generate_embeddings(['test'])
        health_report['traditional_rag']['embedding_manager'] = '✅ Healthy'
    except Exception as e:
        health_report['traditional_rag']['embedding_manager'] = f'❌ Error: {str(e)[:50]}'
    
    try:
        test_response = llm.invoke(['Hello'])
        health_report['traditional_rag']['llm'] = '✅ Healthy'
    except Exception as e:
        health_report['traditional_rag']['llm'] = f'❌ Error: {str(e)[:50]}'
    
    # Check agentic RAG components
    if agentic_rag:
        try:
            agent_count = len(agentic_rag.agents)
            health_report['agentic_rag']['agents'] = f'✅ Healthy ({agent_count} agents)'
        except Exception as e:
            health_report['agentic_rag']['agents'] = f'❌ Error: {str(e)[:50]}'
        
        health_report['agentic_rag']['integration'] = '✅ Successfully integrated'
    else:
        health_report['agentic_rag']['integration'] = '❌ Not integrated'
    
    return health_report

# Run health check
health_status = check_system_health()

print("🏥 SYSTEM HEALTH REPORT")
print("=" * 80)

print("\n🔹 Traditional RAG Components:")
for component, status in health_status['traditional_rag'].items():
    print(f"   {component}: {status}")

print("\n🔹 Agentic RAG Components:")
for component, status in health_status['agentic_rag'].items():
    print(f"   {component}: {status}")
```

## Error Handling Best Practices

Here are some key improvements made to fix the processing_steps error:

### Key Issues Fixed:
1. **Missing Context Initialization**: The processing context wasn't properly initialized with all required keys
2. **Agent Return Values**: Agents weren't properly preserving the context structure
3. **Error Propagation**: Errors weren't being handled gracefully in the agent pipeline

### Best Practices Applied:
1. **Defensive Programming**: Always check if keys exist before accessing them
2. **Context Preservation**: Ensure agents maintain the context structure
3. **Graceful Degradation**: Handle errors without breaking the entire pipeline
4. **Comprehensive Error Responses**: Provide meaningful error messages with full metadata

### Production Tips:
1. Always initialize all required context keys upfront
2. Use `.get()` method with defaults when accessing dictionary keys
3. Implement proper error handling at each agent level
4. Log detailed information for debugging
5. Provide fallback responses when agents fail
